````python
types = [np.int8, np.int16, np.int32, np.int64, np.float16, np.float32, np.float64]
for dtype in types:
    if np.issubdtype(dtype, np.integer):
        info = np.iinfo(dtype)
    else:
        info = np.finfo(dtype)
    print(f"{dtype.__name__}: min={info.min}, max={info.max}")
````

9. Как отсортировать массив по n-му столбцу?

````python
def sortCol(arr, n):
    return arr[arr[:, n].argsort()]
arr = np.random.randint(0, 100, (5, 5))
print("Original:\n", arr)
print("by 2 column:\n", sortCol(arr, 2))
````

10. Как определить, есть ли в данном двумерном массиве пустые столбцы?

````python
def findEmptyColumns(arr):
    emptyСolumns = []
    for col in range(arr.shape[1]):
        if np.all(arr[:, col] == 0):
            emptyСolumns.append(col)
    return emptyСolumns

arr = np.random.randint(0, 2, (3, 5))
arr[:, 2] = 0 
print("Array:\n", arr)
print("Empty columns:", findEmptyColumns(arr))
````

2. Сбалансировано ли соотношение мужчин и женщин в наборе данных?

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
df[['BP_systolic', 'BP_diastolic']] = df['Blood Pressure (s/d)'].str.split('/', expand=True).astype(float)

high_stress = df[df['Stress Levels'] > 70]
low_stress = df[df['Stress Levels'] <= 30]

high_stress_bp = high_stress['BP_systolic'].mean()
low_stress_bp = low_stress['BP_systolic'].mean()

print("True" if high_stress_bp > low_stress_bp else "False")
````

4. Правда ли, что уровень холестерина у мужчин и женщин отличается?

````python
cholesterol_by_gender = df.groupby('Gender')['Cholesterol Level (mg/dL)'].mean()
print("True" if cholesterol_by_gender.Female - cholesterol_by_gender.Male > 10 else "False")
````

5. Правда ли, что с возрастом женщины имеют свойство оценивать свои когнитивные способности хуже, чем в юности (порог юности - 30 лет, женщина в возрасте - 60 лет+)

````python
women = df[df['Gender'] == 'Female']

young_women = women[women['Age (years)'] <= 30]
older_women = women[women['Age (years)'] >= 60]

young_cognitive = young_women['Cognitive Function'].mean()
older_cognitive = older_women['Cognitive Function'].mean()

print("True" if young_cognitive - older_cognitive > 10 else "False")
````

6. Влияет ли физическая активность на вес в возрасте более 50 лет?

````python
over_50 = df[df['Age (years)'] > 50]
weight_by_activity = over_50.groupby('Physical Activity Level')['Weight (kg)'].mean()

max_weight = weight_by_activity.max()
min_weight = weight_by_activity.min()
weight_difference = max_weight - min_weight

print("True" if weight_difference > 5 else "False")
````

7. Сравните минимальный вес мужчин с высоким и низким заработком

````python
high_salary = df[(df["Gender"] == "Male") & (df["Income Level"] == "High")]
low_salary = df[(df["Gender"] == "Male") & (df["Income Level"] == "Low")]

low_salary_fluff = low_salary["Weight (kg)"].min()
high_salary_fluff = high_salary["Weight (kg)"].min()

print("Low salary man has more weight than high salary man" if low_salary_fluff > high_salary_fluff else "High salary man has more weight than low salary man")
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

````python
BMI_more_75 = df[df["BMI"] >= 75]
BMI_30_60 = df[(df["BMI"] >= 30) & (df["BMI"] <= 60)]

count_30_60 = BMI_30_60["Chronic Diseases"].notna().sum()
count_mode_75 = BMI_more_75["Chronic Diseases"].notna().sum()

print("True" if count_30_60 < count_mode_75 else "False")
````

9. Правда ли, что люди с высоким уровнем дохода имеют больший уровень физической активности, чем люди с низким доходом

````python
high_salary = df[df["Income Level"] == "High"]
low_salary = df[df["Income Level"] == "Low"]

count_high_active = (high_salary["Physical Activity Level"] != "Low").sum()
count_low_active = (low_salary["Physical Activity Level"] != "Low").sum()

print("True" if count_high_active > count_low_active else "False")
````

10. Какая доля мужчин с высокой частотой пребывания на солнце имеет высокий заработок?

````python
sun_exposure_median = df["Sun Exposure"].median()
high_sun_exposure = df["Sun Exposure"] > sun_exposure_median

men_high_sun = df[(df["Gender"] == "Male") & high_sun_exposure]

high_income_men = men_high_sun[men_high_sun["Income Level"] == "High"]
proportion = len(high_income_men) / len(men_high_sun)

print(f"The proportion of men with a high frequency of sun exposure and high salaey: {proportion}")
print(f"Number of men with high frequency of sun exposure: {len(men_high_sun)}")
print(f"With high salary: {len(high_income_men)}")
````
